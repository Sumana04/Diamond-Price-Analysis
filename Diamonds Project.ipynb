```python
#To read the data
data=pd.read_csv(r"C:\Users\admin\OneDrive\Desktop\Student\diamonds.csv")
print(data)
```

```python
#Renameing the columns 
data = data.rename(columns={'depth': 'depth(%)', 'x': 'x(length in mm)','y':'y(width in mm)','z':'z(depth in mm)'})
print(data)
```

```python
# converting price column from Dollars to Indian Rupees

# Define the fixed conversion rate
usd_to_inr_rate = 74.50    

# Convert the 'price' column from USD to INR
data['price'] = data['price'] * usd_to_inr_rate

# Save the updated DataFrame back to the same CSV file
data.to_csv('diamonds.csv', index=False)
print(data)
```

```python
Clarity - Measurement of how clear the diamond is.

IF (Internally Flawless):It is Extremely rare and highly valuable. Appears perfect to the naked eye and even under magnification,
       except for some surface characteristics
VVS1 and VVS2 (Very, Very Slightly Included): Inclusions are very difficult to detect under 10x magnification. VVS1 is a higher grade than VVS2.
VS1 and VS2 (Very Slightly Included): Inclusions are minor and range from difficult (VS1) to somewhat easy (VS2) to see under 10x magnification.
SI1 and SI2 (Slightly Included): Inclusions are noticeable under 10x magnification. SI1 has fewer or smaller inclusions compared to SI2.
I1(Included): Inclusions are obvious under 10x magnification and may affect transparency and brilliance. I1 is the best of this category.
```

```python
Color - 
- D: Absolutely colorless. The highest color grade, extremely rare.(Best Quality)
- E: Colorless. Only minute traces of color can be detected by an expert gemologist. A rare and high-quality diamond.
- F: Colorless. Slight color detected by an expert gemologist, but still considered a high-quality diamond.
- G-H: Near-colorless. Color is noticeable when compared to diamonds of better grades, but these diamonds offer excellent value.
- I-J: Near-colorless. Slightly detectable color, but offer excellent value.(Worst Quality).
```

```python
# Analyzing the distribution of diamond prices 
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=50, kde=True,color='m') # kde= Kernel Density Estimate (KDE) 
plt.title('Distribution of Diamond Prices',fontsize=16)
plt.xlabel('Price',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.show()

# Here Frequency depends on price
```

```python
# Analysing correlation among different columns

numeric_df = data.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix HeatMap')
plt.show()

#correlation points lies between -1 and 1
# 1 indicates- perfect positive correlation
# -1 - perfect negative correlation

# Color Intensity represents the strength of the correlation
# Darker Red- Strong positive Correlation
# draker blue-strong negative  orrelation
# light color - weaker correlation
```

```python
# Grouping cut and price column and finding the mean 
data.groupby('cut')['price'].mean()
```

```python
# Carat vs price

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='carat', y='price', alpha=0.5)
plt.title('Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
# Bar plot: Comparison of prices based on cut quality
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='cut', y='price',color="#ADD8E6")
plt.title('Prices based on Cut Quality')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```

```python
#histogram to visualize the distribution of depth values in the diamond
plt.figure(figsize=(12, 6))
sns.histplot(data['depth(%)'], kde=False, bins=30,color="y")
plt.xlabel('Depth (%)')
plt.ylabel('Frequency')
plt.title('Histogram of Depth Values')
plt.show()
```
